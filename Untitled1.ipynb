```python
class StudyTaskPrioritizer:


    def __init__(self):

        self.tasks = []

    def _get_valid_score(self, prompt):

        while True:
            try:
                score = int(input(prompt))
                if 1 <= score <= 5:
                    return score
                else:
                    print(" Score must be an integer between 1 and 5. Please try again.")
            except ValueError:
                print(" Invalid input. Please enter a number.")

    def add_task_interactive(self):

        print("\n-- Add New Task ---")
        name = input("Enter the task description: ")

        importance = self._get_valid_score(
            "Enter Importance (1=Low, 5=High): "
        )

        urgency = self._get_valid_score(
            "Enter Urgency (1=Low, 5=High): "
        )


        priority_score = importance * urgency

        task = {
            'name': name,
            'importance': importance,
            'urgency': urgency,
            'priority_score': priority_score
        }
        self.tasks.append(task)
        print(f" Task '{name}' added with Priority Score: {priority_score}")
        print("-" * 25)

    def prioritize_tasks(self):


        self.tasks.sort(key=lambda task: task['priority_score'], reverse=True)
        print("\n Tasks prioritized successfully!")

    def display_tasks(self):

        if not self.tasks:
            print("\n No tasks currently in the list. Add some study tasks!")
            return

        print("\n---  Prioritized Study Task List ---")
        print("{:<4} | {:<40} | {:<10} | {:<8} | {:<8}".format(
            "Rank", "Task Name", "Importance", "Urgency", "Priority"))
        print("-" * 78)

        for i, task in enumerate(self.tasks, 1):
            print("{:<4} | {:<40} | {:<10} | {:<8} | {:<8}".format(
                i,
                task['name'],
                task['importance'],
                task['urgency'],
                task['priority_score']
            ))
        print("-" * 78)

    def get_top_priority_task(self):

        if not self.tasks:
            return "No tasks in the list."

        self.prioritize_tasks()
        top_task = self.tasks[0]

        print("\n Your **TOP PRIORITY** Task is:")
        print(f"   Name: **{top_task['name']}**")
        print(f"   Priority Score: **{top_task['priority_score']}** (Importance: {top_task['importance']}, Urgency: {top_task['urgency']})")
        print("   -> Focus on this task first!")

    def run_assistant(self):

        while True:
            print("\n--- Time Management Assistant Menu ---")
            print("1. Add a new task")
            print("2. Display prioritized tasks")
            print("3. Show top priority task")
            print("4. Exit")

            choice = input("Enter your choice (1-4): ")

            if choice == '1':
                self.add_task_interactive()
            elif choice == '2':
                self.prioritize_tasks()
                self.display_tasks()
            elif choice == '3':
                self.get_top_priority_task()
            elif choice == '4':
                print(" Thank you for using the Study Task Prioritizer. Goodbye!")
                break
            else:
                print(" Invalid choice. Please enter a number between 1 and 4.")


# --- Example Usage ---
if __name__ == "__main__":
    assistant = StudyTaskPrioritizer()
    assistant.run_assistant()
```
